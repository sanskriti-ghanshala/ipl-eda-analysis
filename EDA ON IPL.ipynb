```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Dell\Downloads\archive (2)\matches.csv")

df.head()
```

```python
df.info()

df.describe()
```

```python
wins = df['match_winner'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=wins.values, y=wins.index)

plt.title("Most Winning IPL Teams")

plt.xlabel("Total Wins")

plt.ylabel("Team")

plt.show()
```

```python
top_players = df['player_of_the_match'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=top_players.values, y=top_players.index)

plt.title("Top Player of the Match Winners")

plt.xlabel("Awards")

plt.ylabel("Player")

plt.show()
```

```python
top_stadiums = df['venue'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=top_stadiums.values, y=top_stadiums.index)

plt.title("Most Used IPL Stadiums")

plt.xlabel("Number of Matches")

plt.ylabel("Stadium")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['margin'], bins=20)

plt.title("Match Winning Margin Distribution")

plt.show()
```
